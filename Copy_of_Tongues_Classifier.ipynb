```python
import numpy as np
import os, sys
import inspect

from sklearn.metrics import classification_report

app_path = inspect.getfile(inspect.currentframe())
sub_dir = os.path.realpath(os.path.dirname(app_path))
main_dir = os.path.dirname(sub_dir)

sys.path.insert(0, main_dir)
%cd /content/drive/MyDrive/Tongue_Classifier
from preprocess import process
from sklearn import preprocessing
import keras
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization


# path for images
'''
Having 2 different categories to identify
Configuring the folders for each class
Configuring the labels for each class
'''
#%cd /content/drive/MyDrive/
TRAINING_FOLDER = os.path.join(sub_dir, 'Tongue_Classifier/train/')
TESTING_FOLDER = os.path.join(sub_dir, 'Tongue_Classifier/test/')
```

```python
train_dateset, train_label = create_training_dataset()
test_dataset, test_label = process_test_dataset()

print(test_label)
# Splitting the feature and label for train dataset
empty_arr = [i for i, x in enumerate(train_dateset) if x == []]
feature_train = [i for i in train_dateset if i != []]
feature_train = np.array(feature_train).reshape(-1, row, cox, band)
labels_train = [i for i in train_label]
labels_train = np.array(labels_train).astype(str)

# Splitting the feature and label for test dataset
# feature_test = [i for i in test_dataset]
feature_test = [i for i in test_dataset if i != []]
feature_test = np.array(feature_test).reshape(-1, row, cox, band)
labels_test = [i for i in test_label]
labels_test = np.array(labels_test).astype(str)

number = preprocessing.LabelEncoder()

labels_train = number.fit_transform(labels_train)
labels_test = number.fit_transform(labels_test)

######################
y_train = keras.utils.to_categorical(labels_train)
y_test = keras.utils.to_categorical(labels_test)
```

```python
print(feature_train.shape)
print(feature_test.shape)
print(y_train.shape)
print(y_test.shape)
n_classes = y_train.shape[1]
```

```python
def larger_model():
    # create model
    model = Sequential()
    model.add(Conv2D(10, (3, 3), input_shape=(row, cox, band), activation='relu'))
    model.add(BatchNormalization())
    model.add(Conv2D(15, (3, 3), activation='relu'))
    model.add(BatchNormalization())
    model.add(Conv2D(20, (3, 3), activation='relu'))
    model.add(BatchNormalization())
    model.add(Conv2D(25, (3, 3), activation='relu'))
    model.add(BatchNormalization())
    model.add(Conv2D(30, (3, 3), activation='relu'))
    model.add(BatchNormalization())
    model.add(MaxPooling2D())
    model.add(Flatten())
    model.add(Dense(4096, activation='relu'))
    model.add(Dropout(0.4))
    model.add(Dense(1058, activation='relu'))
    model.add(Dense(n_classes, activation='softmax'))
    # Compile model
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model



# Fit the model
model = larger_model()
model.fit(feature_train, y_train, validation_split=0.20, epochs=3000, batch_size=200, shuffle=True)
# Final evaluation of the model
# scores = model.evaluate(feature_test, y_test, verbose=0)
scores = model.predict(feature_test)
print(y_test)
print(scores)
```

```python
# print("CNN Accuracy: %.2f%%" % (scores[1] * 100))

model.save(os.path.join(main_dir, '/content/drive/MyDrive/Tongue_Classifier/trained_model/model.h5'))


Y_test = np.argmax(y_test, axis=1) # Convert one-hot to index
out = np.argmax(scores.round(), axis=1)
print(classification_report(Y_test, out))
```
